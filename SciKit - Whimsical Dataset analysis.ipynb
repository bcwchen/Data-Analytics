```python
#Â© Anaconda Inc. 2018

import pandas as pd
import warnings
warnings.simplefilter('ignore')
```

```python
filename='Learning about Humans learning ML.csv'
human = pd.read_csv(filename)
```

```python
human.head(5)
```

```python
human.drop(['Timestamp'],1,inplace=True)
human['Education']=(human['Years of post-secondary education (e.g. BA=4; Ph.D.=10)']
                    .str.replace(r'.*=','')
                    .astype(int)
                   )
human.drop(['Years of post-secondary education (e.g. BA=4; Ph.D.=10)'], 
            1, inplace=True)
```

```python
human.head(5)
```

```python
human.describe(include=['object','int','float'])
```

```python
human_dummies = pd.get_dummies(human)
list(human_dummies.columns)
```

```python
X = human_dummies.drop(['How successful has this tutorial been so far?'],1)
y = human_dummies['How successful has this tutorial been so far?']>=8   
```

```python
y.head()
```

```python
X.iloc[:5,:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
print 'Training features/target:',X_train.shape, y_train.shape
print 'Testing features/target:', X_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10,random_state=0)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=7, random_state=0)
tree.fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
tree = DecisionTreeClassifier(max_depth=7, random_state=0)
tree.fit(X, y)
tree.score(X, y)
```

```python
%matplotlib inline
pd.Series(tree.feature_importances_, index=X.columns).plot.barh(figsize=(18,7));
```

```python
from sklearn.tree import export_graphviz
import sys, subprocess
from IPython.display import Image
```

```python
export_graphviz(tree, feature_names=X.columns, class_names=['failure','success'],
                out_file='ml-good.dot', impurity=False, filled=True)
```

```python
subprocess.check_call([sys.prefix+'/bin/dot','-Tpng','ml-good.dot',
                       '-o','ml-good.png'])
Image('ml-good.png')
```
